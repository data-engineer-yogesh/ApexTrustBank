### Tables Design 

Core Tables 
1.  customers
2.  accounts
3.  transactions
4.  branches


1. customers column 

    1. customer_id (PK)
    2. name
    3. city
    4. kyc_status  (VERIFIED / PENDING)
    5. created_date

2. accounts
    1.  account_id (PK)
    2.  customer_id (FK)
    3.  account_type (SAVINGS / CURRENT)
    4.  balance
    5.  opened_date


3. transactions

      1.   txn_id (PK)
      2.  account_id (FK)
      3.   txn_type (DEBIT / CREDIT)
      4.   amount
      5.  txn_date



4.branches

      1.  branch_id
      2.  branch_city
      3.  region
