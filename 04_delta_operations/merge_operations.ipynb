### Merge Operations

```sql

CREATE OR REPLACE TABLE apextrust_catalog.bronz.cdc_customers 
AS SELECT * FROM apextrust_catalog.silver.customers;
```

```sql
-- Use correct catalog and schema
USE CATALOG apextrust_catalog;
USE SCHEMA silver;

-- MERGE CDC data into Silver Customers
MERGE INTO apextrust_catalog.silver.customers AS target
USING apextrust_catalog.bronz.cdc_customers AS source
ON target.customer_id = source.customer_id

WHEN MATCHED THEN
  UPDATE SET
    target.name = source.name,
    target.city = source.city,
    target.kyc_status = source.kyc_status,
    target.created_date = source.created_date

WHEN NOT MATCHED THEN
  INSERT (
    customer_id,
    name,
    city,
    kyc_status,
    created_date,
    silver_processed_ts
  )
  VALUES (
    source.customer_id,
    source.name,
    source.city,
    source.kyc_status,
    source.created_date,
    current_timestamp()
  );
```

- MERGE INTO = upsert
- Supported only by Delta tables
- Very common scenario question
