### Optimize vacuum  
( Performance + Cost)
  Purpose
-       Improve query performance
-       Clean old files
-       Reduce storage cost


```sql
USE CATALOG apextrust_catalog;
```

```sql
-- Optimize Gold tables (file compaction)

OPTIMIZE apextrust_catalog.gold.balance_by_city;
OPTIMIZE apextrust_catalog.gold.daily_transactions;
```

```sql
-- Z-Ordering (optional but advanced)

OPTIMIZE apextrust_catalog.gold.daily_transactions
ZORDER BY (txn_date);
```

```sql
-- View files before cleanup (optional)

DESCRIBE DETAIL apextrust_catalog.gold.daily_transactions;
```

```sql
-- Vacuum old files (retain 7 days - SAFE)

VACUUM apextrust_catalog.gold.daily_transactions RETAIN 168 HOURS;
```


- OPTIMIZE = performance
- VACUUM = storage cleanup
- Default retention = 7 days
- Disabling retention =  dangerous
