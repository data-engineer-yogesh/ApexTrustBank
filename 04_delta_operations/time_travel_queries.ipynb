#### Time Travel Queries 
(Regulatory / Audit Favorite)

Purpose
- Query historical versions
- Audit changes
- Rollback mistakes


```sql
USE CATALOG apextrust_catalog;
```

```sql
-- View table history
DESCRIBE HISTORY apextrust_catalog.silver.customers;
```

```sql
-- Query by VERSION
SELECT *
FROM apextrust_catalog.silver.customers VERSION AS OF 1;
```

```sql
-- Query by TIMESTAMP

SELECT *
FROM apextrust_catalog.silver.customers
TIMESTAMP AS OF '2026-01-19T09:52:53.068+00:00';
```

```sql
-- Restore table to previous version (DANGEROUS but exam-relevant)

RESTORE TABLE apextrust_catalog.silver.customers TO VERSION AS OF 1;
```


- Time Travel = Delta Lake
- Used for audits, rollback
- Cannot be done on CSV/Parquet
